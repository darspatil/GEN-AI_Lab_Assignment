```python
# Cell 2: Load the QA Pipeline
from transformers import pipeline

# Initialize the pipeline
# This automatically downloads the model (~260MB)
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

print("QA Pipeline loaded.")
```

```python
# Cell 3: Define Context
context = """
Python is a high-level, general-purpose programming language. Its design philosophy emphasizes code readability with the use of significant indentation.
Python is dynamically typed and garbage-collected. It supports multiple programming paradigms, including structured (particularly procedural), object-oriented and functional programming.
Guido van Rossum began working on Python in the late 1980s as a successor to the ABC programming language and first released it in 1991 as Python 0.9.0.
Python consistently ranks as one of the most popular programming languages.
"""

print("Context loaded.")
```

```python
# Cell 4: Define the Chat Function
def get_answer(question, context):
    result = qa_pipeline(question=question, context=context)
    return result['answer'], result['score']

# Test it once
q = "Who created Python?"
ans, score = get_answer(q, context)
print(f"Q: {q}")
print(f"A: {ans} (Confidence: {score:.2f})")
```

```python
# Cell 5: Interactive Chatbot
print("--- Simple QA Chatbot ---")
print("Context: Python History & Features (See Cell 3)")
print("Type 'exit' or 'quit' to stop.\n")

while True:
    # 1. Get user input
    user_input = input("Your Question: ")

    # 2. Check for exit command
    if user_input.lower() in ['exit', 'quit']:
        print("Goodbye!")
        break

    # 3. Get answer from the model
    # We pass the same 'context' every time. In a real app, you might retrieve different contexts.
    try:
        answer, confidence = get_answer(user_input, context)

        # 4. Filter out low confidence answers (Optional)
        if confidence < 0.1:
            print("Chatbot: I am not sure about that based on the text provided.")
        else:
            print(f"Chatbot: {answer}\n")

    except Exception as e:
        print(f"An error occurred: {e}")
```
